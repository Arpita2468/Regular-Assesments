```python
pu=euro['Foot'].dropna().unique().tolist()
pu.remove("-")
pu
```

```python
import random as r
def moon(a):
    if isnull(a):
        return r.choice(pu)
    elif a=="-":
        return r.choice(pu) 
    else:
        return a
euro['Foot']=euro['Foot'].apply(moon)
euro.tail()
```
